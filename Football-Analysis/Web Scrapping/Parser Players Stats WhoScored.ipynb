# Imports

```python
from selenium import webdriver
import pandas as pd
import time
```

# Definition des parsers

```python
def parse_sum(n):
    headers_table = []
    final_results = []
    
    All_players_button =driver.find_elements_by_xpath('/html/body/div[6]/div[2]/div[2]/div[5]/div[2]/div[1]/div[2]/div[2]/dl/dd[2]/a')[0]
    All_players_button.click()
    
    rows = driver.find_elements_by_css_selector("#top-player-stats-summary-grid")
    
#Recuperation des headers
    for row in rows:
        temp = []
        columns = row.find_elements_by_css_selector("th")
        for column in columns:
            temp.append(column.text)
        headers_table.append(temp)
    
#Recuperation des results:
    for i in range(n):
        
        rows = driver.find_elements_by_css_selector("#top-player-stats-summary-grid")
        results_table = []
        
        for row in rows:
            temp = []
            columns = row.find_elements_by_css_selector("td")
            for column in columns:
                temp.append(column.text)
            results_table.append(temp)
            results_table_final=[results_table[0][i*14:14*(i+1)] for i in range(10)]
        final_results += results_table_final
        next_button =driver.find_elements_by_xpath('/html/body/div[6]/div[2]/div[2]/div[5]/div[2]/div[4]/div/dl[2]/dd[3]/a')[0]
        next_button.click()
        time.sleep(1)
        
    return headers_table,final_results
```

```python
def parse_def(n):
    headers_table = []
    final_results = []
    
    Defense_button =driver.find_elements_by_xpath('/html/body/div[6]/div[2]/div[2]/div[5]/ul/li[2]/a')[0]
    Defense_button.click()
    
    time.sleep(1)
    
    All_players_button =driver.find_elements_by_xpath('/html/body/div[6]/div[2]/div[2]/div[5]/div[3]/div[1]/div[2]/div[2]/dl/dd[2]/a')[0]
    All_players_button.click()
    
    rows = driver.find_elements_by_css_selector("#statistics-table-defensive")
    
#Recuperation des headers
    for row in rows:
        temp = []
        columns = row.find_elements_by_css_selector("th")
        for column in columns:
            temp.append(column.text)
        headers_table.append(temp)
    
#Recuperation des results:
    for i in range(n):
        
        rows = driver.find_elements_by_css_selector("#statistics-table-defensive")
        results_table = []
        
        for row in rows:
            temp = []
            columns = row.find_elements_by_css_selector("td")
            for column in columns:
                temp.append(column.text)
            results_table.append(temp)
            results_table_final=[results_table[0][i*14:14*(i+1)] for i in range(10)]
        final_results += results_table_final
        next_button =driver.find_elements_by_xpath('/html/body/div[6]/div[2]/div[2]/div[5]/div[3]/div[4]/div/dl[2]/dd[3]/a')[0]
        next_button.click()
        time.sleep(1)
        
    return headers_table,final_results
```

```python
def parse_off(n):
    headers_table = []
    final_results = []
    
    Offensive_button =driver.find_elements_by_xpath('/html/body/div[6]/div[2]/div[2]/div[5]/ul/li[3]/a')[0]
    Offensive_button.click()
    
    time.sleep(1)
    
    All_players_button =driver.find_elements_by_xpath('/html/body/div[6]/div[2]/div[2]/div[5]/div[4]/div[1]/div[2]/div[2]/dl/dd[2]/a')[0]
    All_players_button.click()
    
    rows = driver.find_elements_by_css_selector("#statistics-table-offensive")
    
#Recuperation des headers
    for row in rows:
        temp = []
        columns = row.find_elements_by_css_selector("th")
        for column in columns:
            temp.append(column.text)
        headers_table.append(temp)
    
#Recuperation des results:
    for i in range(n):
        
        rows = driver.find_elements_by_css_selector("#statistics-table-offensive")
        results_table = []
        
        for row in rows:
            temp = []
            columns = row.find_elements_by_css_selector("td")
            for column in columns:
                temp.append(column.text)
            results_table.append(temp)
            results_table_final=[results_table[0][i*15:15*(i+1)] for i in range(10)]
        final_results += results_table_final
        next_button =driver.find_elements_by_xpath('/html/body/div[6]/div[2]/div[2]/div[5]/div[4]/div[4]/div/dl[2]/dd[3]/a')[0]
        next_button.click()
        time.sleep(1.5)
        
    return headers_table,final_results
```

```python
def parse_pass(n):
    headers_table = []
    final_results = []
    
    Passing_button =driver.find_elements_by_xpath('//*[@id="stage-top-player-stats-options"]/li[4]/a')[0]
    Passing_button.click()
    
    time.sleep(1)
    
    All_players_button =driver.find_elements_by_xpath('/html/body/div[6]/div[2]/div[2]/div[5]/div[5]/div[1]/div[2]/div[2]/dl/dd[2]/a')[0]
    All_players_button.click()
    
    rows = driver.find_elements_by_css_selector("#statistics-table-passing")
    
#Recuperation des headers
    for row in rows:
        temp = []
        columns = row.find_elements_by_css_selector("th")
        for column in columns:
            temp.append(column.text)
        headers_table.append(temp)
    
#Recuperation des results:
    for i in range(n):
        
        rows = driver.find_elements_by_css_selector("#statistics-table-passing")
        results_table = []
        
        for row in rows:
            temp = []
            columns = row.find_elements_by_css_selector("td")
            for column in columns:
                temp.append(column.text)
            results_table.append(temp)
            results_table_final=[results_table[0][i*13:13*(i+1)] for i in range(10)]
        final_results += results_table_final
        next_button =driver.find_elements_by_xpath('/html/body/div[6]/div[2]/div[2]/div[5]/div[5]/div[4]/div/dl[2]/dd[3]/a')[0]
        next_button.click()
        time.sleep(2)
        
    return headers_table,final_results
```

# Nettoyage de la table finale

```python
# dropping null value columns to avoid errors 
df.dropna(inplace = True) 
  
# new data frame with split value columns 
new = df['Player'].str.split("\n", n = 1, expand = True) 
new2 = new[1].str.split(",", n = 1, expand = True)
new3 = new2[1].str.split(",", n = 1, expand = True)
new4 = new3[1].str.split(",", n = 1, expand = True)
#new5 = new4[1].str.split(",", n = 1, expand = True)
new6 =  df['Apps'].str.split("(", n = 1, expand = True) 
new7 = new6[1].str.split(")", n = 1, expand = True)

df['Name']= new[0] 
df['Team']= new2[0] 
df['Age']= new3[0] 
df['Position 1']= new4[0] 
df['Position 2']= new4[1] 
#df['Position 3']= new5[1] 
df['Games Started']= new6[0] 
df['Games Subbed In']= new7[0] 

df = df.drop(columns=['Player', 'Rating', 'Apps', ''])
```
