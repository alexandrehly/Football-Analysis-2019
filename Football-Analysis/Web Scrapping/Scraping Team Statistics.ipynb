# Definition des parsers

```python
def parse_glob(n,teams,parameters,table,button="None"):
    headers_table = []
    final_results = []
    
    #print(button)
    
    if button!="None":
        
        button_to_click =driver.find_elements_by_xpath(button)[0]
        button_to_click.click()
    
        time.sleep(1)
    
    rows = driver.find_elements_by_css_selector(table)

#Recuperation des headers
    for row in rows:
        temp = []
        columns = row.find_elements_by_css_selector("th")

        for column in columns:
            temp.append(column.text)
        headers_table.append(temp)
    
#Recuperation des results:

    rows = driver.find_elements_by_css_selector(table)
    results_table = []
        
    for row in rows:
        temp = []
        columns = row.find_elements_by_css_selector("td")

        for column in columns:
            temp.append(column.text)
        results_table.append(temp)
        
        results_table_final=[results_table[0][i*parameters:parameters*(i+1)] for i in range(teams)]
        
        final_results += results_table_final
        
    return headers_table,final_results
```
